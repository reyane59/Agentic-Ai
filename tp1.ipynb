```python
tokenizer = tiktoken.get_encoding("o200k_base")
```

```python
prompt = "tu es un expert en mathématiques"
```

```python
tokens = tokenizer.encode(prompt)
```

```python
print(len(tokens))
```

```python
for token in tokens : 
    t= tokenizer.decode_single_token_bytes(token).decode("utf-8")
    print(t)
```

```python
def token_count (prompt:str, model:str="o200k_base"):
    tokenizer = tiktoken.get_encoding(model)
    return len(tokenizer.encode(prompt))
```

```python
print(token_count(prompt="vous êtes un expert en mathématiques"))
```
